```python
# Required Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'RTA Data 2020 to July 2023.xlsx'  # Update path as needed
df = pd.read_excel(file_path)

# Data Preprocessing
df_cleaned = df.drop(columns=['EcYear', 'EcNumber', 'CallTime', 'EmergencyArea', 'HospitalName', 'Reason', 'EducationTitle', 'Cause'])
df_cleaned = df_cleaned.dropna()  # Drop rows with missing values

# Encoding categorical variables
label_encoders = {}
categorical_cols = ['Gender', 'InjuryType', 'PatientStatus']
for col in categorical_cols:
    le = LabelEncoder()
    df_cleaned[col] = le.fit_transform(df_cleaned[col])
    label_encoders[col] = le

# Defining features and targets
X = df_cleaned.drop(columns=['InjuryType', 'PatientStatus'])  # Features
y_injury = df_cleaned['InjuryType']  # Target 1: InjuryType
y_patient = df_cleaned['PatientStatus']  # Target 2: PatientStatus

# Splitting the data into training and testing sets
X_train_injury, X_test_injury, y_train_injury, y_test_injury = train_test_split(X, y_injury, test_size=0.2, random_state=42)
X_train_patient, X_test_patient, y_train_patient, y_test_patient = train_test_split(X, y_patient, test_size=0.2, random_state=42)

# Logistic Regression Model Training and Evaluation

# Function to train and evaluate model
def train_and_evaluate(X_train, X_test, y_train, y_test, solver, max_iter):
    model = LogisticRegression(solver=solver, max_iter=max_iter)
    model.fit(X_train, y_train)

    # Predictions
    y_pred = model.predict(X_test)

    # Evaluation Metrics
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)

    # For multiclass classification, use appropriate 'average' setting
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    return accuracy, conf_matrix, precision, recall, f1


# Testing different solvers and iterations
solvers = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
max_iters = [50, 100, 150, 200, 250, 300]

# Storing results
solver_acc = []
iter_acc = []

# Evaluating InjuryType prediction
for solver in solvers:
    accuracy, conf_matrix, precision, recall, f1 = train_and_evaluate(X_train_injury, X_test_injury, y_train_injury, y_test_injury, solver, 100)
    solver_acc.append(accuracy)
    print(f"Solver: {solver} -> Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1}")

# Evaluating effect of max_iter
for max_iter in max_iters:
    accuracy, conf_matrix, precision, recall, f1 = train_and_evaluate(X_train_injury, X_test_injury, y_train_injury, y_test_injury, 'lbfgs', max_iter)
    iter_acc.append(accuracy)
    print(f"Max_iter: {max_iter} -> Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1}")

# Plotting Accuracy vs Solver
plt.figure(figsize=(10,5))
plt.plot(solvers, solver_acc, marker='o')
plt.title('Accuracy vs Solver (InjuryType)')
plt.xlabel('Solver')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()

# Plotting Accuracy vs Max_iter
plt.figure(figsize=(10,5))
plt.plot(max_iters, iter_acc, marker='o')
plt.title('Accuracy vs Max_iter (InjuryType)')
plt.xlabel('Max_iter')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()

# Repeat for PatientStatus if required
```
